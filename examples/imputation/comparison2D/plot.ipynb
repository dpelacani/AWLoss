```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bilinear = np.load("./bilinear_3_1.npy")
bicubic = np.load("./bicubic_3_1.npy")
mse = np.load("./mse_3_1.npy")
awloss = np.load("./awloss_3_1.npy")
masked = np.load("./masked_3_1.npy")
target = np.load("./target_3_1.npy")

print(bilinear.shape)
print(bicubic.shape)
print(mse.shape)
print(awloss.shape)
print(masked.shape)
print(target.shape)

data = [masked,
        bilinear,
        # bicubic,
        mse,
        awloss,
        target]
labels = [ "Masked Input",
        "Bilinear",
        # "Bicubic",
        "UNet MSE",
        "UNet AWLoss",
        "Ground Truth"]
```

```python
fig, axs  = plt.subplots(1, len(data), figsize=(15,10))
for i in range(len(axs)):
    im = axs[i].imshow(data[i])
    axs[i].set_title(labels[i])
plt.savefig("./fullscale", dpi=500)
plt.show()

```

```python
xi, xf = 40, 110
yi, yf = 40, 110

data_cropped = [d[xi:xf,yi:yf] for d in data]

fig, axs  = plt.subplots(1, len(data_cropped), figsize=(15,10))
for i in range(len(axs)):
    im = axs[i].imshow(data_cropped[i])
    axs[i].set_title(labels[i])
# plt.savefig("./cropped", dpi=500)
plt.show()

```
